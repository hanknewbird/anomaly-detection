```python
from model_plus import NewNet
from PIL import Image
import os
import torch
from torchvision.transforms import Compose, Grayscale, ToTensor
import matplotlib.pyplot as plt
```

```python
# 模型路径
model_Class = "Class8"
model_path = f"./model/NewNet_model_{model_Class}.pt"
model_thresholded = 0.005

# 使用模型
model = NewNet()
model.load_state_dict(torch.load(model_path))
model.eval()
model = model.to('cuda')
transform = Compose([Grayscale(), ToTensor()])

imgpath = f"./image/{model_Class}"
imglist = ["img1.PNG","img2.PNG","img3.PNG","img4.PNG",]

plt.figure(figsize=(15, 10))

img1 = Image.open(os.path.join(imgpath,imglist[0])).convert('L')
img1 = transform(img1)
img1 = img1.to('cuda')
img1 = img1.unsqueeze(0)
y1 = model(img1)
residual1 = torch.abs(img1[0][0]-y1[0][0])

plt.subplot(241)
plt.imshow(img1.detach().cpu().numpy()[0][0],cmap=plt.cm.gray)
plt.title('Image')
plt.axis('off')

plt.subplot(245)
plt.imshow(residual1.detach().cpu().numpy()>model_thresholded,cmap=plt.cm.gray)
plt.title(f"Residual Thresholded={model_thresholded}")
plt.axis('off')

img2 = Image.open(os.path.join(imgpath,imglist[1])).convert('L')
img2 = transform(img2)
img2 = img2.to('cuda')
img2 = img2.unsqueeze(0)
y2 = model(img2)
residual2 = torch.abs(img2[0][0]-y2[0][0])

plt.subplot(242)
plt.imshow(img2.detach().cpu().numpy()[0][0],cmap=plt.cm.gray)
plt.title('Image')
plt.axis('off')

plt.subplot(246)
plt.imshow(residual2.detach().cpu().numpy()>model_thresholded,cmap=plt.cm.gray)
plt.title(f"Residual Thresholded={model_thresholded}")
plt.axis('off')


img3 = Image.open(os.path.join(imgpath,imglist[2])).convert('L')
img3 = transform(img3)
img3 = img3.to('cuda')
img3 = img3.unsqueeze(0)
y3 = model(img3)
residual3 = torch.abs(img3[0][0]-y3[0][0])

plt.subplot(243)
plt.imshow(img3.detach().cpu().numpy()[0][0],cmap=plt.cm.gray)
plt.title('Image')
plt.axis('off')

plt.subplot(247)
plt.imshow(residual3.detach().cpu().numpy()>model_thresholded,cmap=plt.cm.gray)
plt.title(f"Residual Thresholded={model_thresholded}")
plt.axis('off')


img4 = Image.open(os.path.join(imgpath,imglist[3])).convert('L')
img4 = transform(img4)
img4 = img4.to('cuda')
img4 = img4.unsqueeze(0)
y4 = model(img4)
residual4 = torch.abs(img4[0][0]-y4[0][0])

plt.subplot(244)
plt.imshow(img4.detach().cpu().numpy()[0][0],cmap=plt.cm.gray)
plt.title('Image')
plt.axis('off')

plt.subplot(248)
plt.imshow(residual4.detach().cpu().numpy()>model_thresholded,cmap=plt.cm.gray)
plt.title(f"Residual Thresholded={model_thresholded}")
plt.axis('off')

# plt.tight_layout(pad=0.2)
plt.tight_layout(pad=0.4, w_pad=0.8, h_pad=0.1)

plt.show()
```
