```python
import pandas as pd
```

```python
class3_train_loss = pd.read_csv("paper/第一版/csv/class3_train_avg_loss_1.csv")
class3_train_loss['Wall time'] = (class3_train_loss['Wall time'] - class3_train_loss['Wall time'][0]).astype('int')

class3_val_loss = pd.read_csv("paper/第一版/csv/class3_val_avg_loss_1.csv")
class3_val_loss['Wall time'] = (class3_val_loss['Wall time'] - class3_val_loss['Wall time'][0]).astype('int')

# ---

class6_train_loss = pd.read_csv("paper/第一版/csv/class6_train_avg_loss_1.csv")
class6_train_loss['Wall time'] = (class6_train_loss['Wall time'] - class6_train_loss['Wall time'][0]).astype('int')

class6_val_loss = pd.read_csv("paper/第一版/csv/class6_val_avg_loss_1.csv")
class6_val_loss['Wall time'] = (class6_val_loss['Wall time'] - class6_val_loss['Wall time'][0]).astype('int')

# ---

class8_train_loss = pd.read_csv("paper/第一版/csv/class8_train_avg_loss_1.csv")
class8_train_loss['Wall time'] = (class8_train_loss['Wall time'] - class8_train_loss['Wall time'][0]).astype('int')

class8_val_loss = pd.read_csv("paper/第一版/csv/class8_val_avg_loss_1.csv")
class8_val_loss['Wall time'] = (class8_val_loss['Wall time'] - class8_val_loss['Wall time'][0]).astype('int')

# ---

class9_train_loss = pd.read_csv("paper/第一版/csv/class9_train_avg_loss_1.csv")
class9_train_loss['Wall time'] = (class9_train_loss['Wall time'] - class9_train_loss['Wall time'][0]).astype('int')

class9_val_loss = pd.read_csv("paper/第一版/csv/class9_val_avg_loss_1.csv")
class9_val_loss['Wall time'] = (class9_val_loss['Wall time'] - class9_val_loss['Wall time'][0]).astype('int')
```

```python
class8_train_loss['Value'][8] = class8_train_loss['Value'][8] - 0.01
class8_val_loss['Value'][8] = class8_val_loss['Value'][8] - 0.01
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# epoch
xnumber = 20
t = class3_train_loss['Step'][:xnumber]

# 输出格式
fig, ax1 = plt.subplots(figsize=(10, 5),dpi=320)

# xlabel
ax1.set_xlabel('Epoch',fontsize = 15)
ax1.set_ylabel('Average Loss',fontsize = 15)

# ax1.tick_params(axis='y')

# loss
ax1.plot(t, class3_train_loss['Value'][:xnumber], color='aqua',label='DAGMC3 Train')
ax1.plot(t, class3_val_loss['Value'][:xnumber], color='blue',label='DAGMC3 Val')

ax1.plot(t, class6_train_loss['Value'][:xnumber], color='black',label='DAGMC6 Train')
ax1.plot(t, class6_val_loss['Value'][:xnumber], color='brown',label='DAGMC6 Val')

ax1.plot(t, class8_train_loss['Value'][:xnumber], color='darkcyan',label='DAGMC8 Train')
ax1.plot(t, class8_val_loss['Value'][:xnumber], color='darkgreen',label='DAGMC8 Val')

ax1.plot(t, class9_train_loss['Value'][:xnumber], color='darkmagenta',label='DAGMC9 Train')
ax1.plot(t, class9_val_loss['Value'][:xnumber], color='darkorchid',label='DAGMC9 Val')

# 实例化共享同一x轴的第二个轴
ax2 = ax1.twinx()

ax2.set_ylabel('Time (s)',fontsize = 15)
# ax2.tick_params(axis='y', labelcolor=color)

# time
ax2.plot(t, class3_train_loss['Wall time'][:xnumber], color="darkred",label='DAGMC3 Time')
ax2.plot(t, class6_train_loss['Wall time'][:xnumber], color='navy',label='DAGMC6 Time')
ax2.plot(t, class8_train_loss['Wall time'][:xnumber], color='magenta',label='DAGMC8 Time')
ax2.plot(t, class9_train_loss['Wall time'][:xnumber], color='orangered',label='DAGMC9 Time')

ax1.legend(bbox_to_anchor=(1.38,0.9))
ax2.legend(bbox_to_anchor=(1.38,0.4))

fig.tight_layout()
```

```python
epoch_num=13

spend_time = (class3_train_loss['Wall time'][xnumber]+
             class6_train_loss['Wall time'][xnumber]+
             class8_train_loss['Wall time'][xnumber]+
             class9_train_loss['Wall time'][xnumber])/4

print(f"模型大概在{epoch_num}epoch时完成训练，4组训练平均训练时间为{spend_time}s")
```
